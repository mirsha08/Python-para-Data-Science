```python
#para que no se tapen unos a otros. le ponemos pasar una lista de columnas.
#creamos una lista, de momento con 2 elementos:

#la lista es con[] así-->[peliculas[peliculas.Genero == 'Accion'].PresupuestoMillones], peliculas[peliculas.Genero == 'Drama'].PresupuestoMillones]

plt.hist([peliculas[peliculas.Genero == 'Accion'].PresupuestoMillones,\
          peliculas[peliculas.Genero == 'Drama'].PresupuestoMillones], \
         bins = 15, stacked=True)
#stacked es para apilar los datos
plt.show()
```

```python
#podemos añadir el resto de generos igual pero es muy farragoso y queda muy manual todo, así que usando ciclos:
# lo que queremos es cambiar el genero por los valores de Genero
#para sacar los valores únicos de Genero --> peliculas.Genero.cat.categories nos devuelve una lista de los valoes

for gen in peliculas.Genero.cat.categories:
    print(gen)

```

```python
#vemos que funciona la lista y recorre las categorias
```

```python
#seguimos, la lista la tiene que crear dentro de la linea vacia, creamos una lista vacia lista1
lista1 = list()
for gen in peliculas.Genero.cat.categories:
    lista1.append(peliculas[peliculas.Genero == gen].PresupuestoMillones)

print(lista1)  #nos da una columna con indice de la pelicula y otra con el presupuesto, y luego nos da otra segunda lista (una por genero)
```

```python
h = plt.hist(lista1, bins=30, stacked=True, rwidth=0.8, )
#histtype : {'bar', 'barstacked', 'step', 'stepfilled'}, default: 'bar' cambia el estilo
```

```python
#para añadir leyendas creamos una nueva lista de etiquetas
mis_etiquetas = list()
for gen in peliculas.Genero.cat.categories:
    lista1.append(peliculas[peliculas.Genero == gen].PresupuestoMillones)
    mis_etiquetas.append(gen)

h = plt.hist(lista1, bins=30, stacked=True, rwidth=0.8, label=mis_etiquetas )
plt.legend()  #muestra las label
plt.ylabel('Presupuesto en Millones')
plt.show() #quita la linea de comando que sale al generar el grafico
```

Para los valores de cmap que es colormapentan linealmente.

- **autum** tonos secuenciales que aumentan linealmente de rojo-naranja-amarillo.  
 - **bone** Mapa de color blanco y negro en aumento secuencial con un matiz de azul, para emular una película de rayos X.    
 - **cool** tonos de cian-magenta que disminuyen linealmente.
 - **copper** tonos crecientes secuenciales de cobre negro.
 - **flag** patrón repetitivo rojo-blanco-azul-negro (no cíclico en los puntos finales).
 - **gray** escala de grises de blanco a negro de aumento lineal secuencial.
 - **hot** secuencial negro-rojo-amarillo-blanco, para emular la radiación del cuerpo negro de un objeto a temperaturas crecientes.
 - **hsv** cíclico rojo-amarillo-verde-cian-azul-magenta-rojo, formado al cambiar el componente de tono en el espacio de color HSV.
 - **inferno** tonos perceptualmente uniformes de negro-rojo-amarillo.
 - **jet** un mapa espectral con puntos finales oscuros, azul-cian-amarillo-rojo; basado en una simulación de chorro de fluido de NCSA..
 - **magma** tonos perceptualmente uniformes de negro-rojo-blanco.
 - **pink** Creciente secuencial en colores pastel negro-rosa-blanco, destinado a la coloración de fotografías en tono sepia.
 - **plasma** tonos perceptualmente uniformes de azul-rojo-amarillo.
 - **prism** patrón repetitivo rojo-amarillo-verde-azul-violeta -…- verde (no cíclico en los puntos finales).
 - **spring** tonos de amarillo magenta que aumentan linealmente.
 - **summer** tonos secuenciales de verde-amarillo que aumentan linealmente.
 - **viridis** tonos perceptualmente uniformes de azul-verde-amarillo.
 - **winter** tonos de azul verdoso que aumentan linealmente.

```python
#creamos subplots que es un paquete de pyplot(plt)
f, axes = plt.subplots(1,2, figsize=(12,6))  #indica 1 fila y 2 columnas, así podemos tener varios graficos para montar
#figsize hace el tamaño  de los graficos
k1= sns.kdeplot(x=peliculas.PresupuestoMillones, y=peliculas.RatingsAudiencia, cmap='plasma', ax =axes[0])
k1= sns.kdeplot(x=peliculas.PresupuestoMillones, y=peliculas.RatingsCriticos, cmap='viridis', ax =axes[1])

#con el ax=axes estamos dando la posición que tiene cada gráfica
```

```python
axes
```

```python
#cuando tenemos más cuadriculas controlamos el axes como en las matrices.
f, axes = plt.subplots(2,2, figsize=(12,6))  
k1= sns.kdeplot(x=peliculas.PresupuestoMillones, y=peliculas.RatingsAudiencia, cmap='plasma', ax =axes[0,1])
k2= sns.kdeplot(x=peliculas.PresupuestoMillones, y=peliculas.RatingsCriticos, cmap='viridis', ax =axes[1,0])
#con el ax=axes estamos dando la posición que tiene cada gráfica
```

```python
#para tener el mismo eje de las x, y para los dos gráficos
f, axes = plt.subplots(1,2, figsize=(12,6), sharex=True, sharey=True) 
k1= sns.kdeplot(x=peliculas.PresupuestoMillones, y=peliculas.RatingsAudiencia, cmap='plasma', ax =axes[0])
k2= sns.kdeplot(x=peliculas.PresupuestoMillones, y=peliculas.RatingsCriticos, cmap='viridis', ax =axes[1])
k1.set(xlim=(-50,250))
```

```python
#Diagramas de Violín
```

```python
z = sns.violinplot(data=peliculas, x = 'Genero', y='RatingsCriticos')
```

```python
#Grafico con Boxplot 
z = sns.boxplot(data=peliculas, x = 'Genero', y='RatingsCriticos')
```

```python
f, axes = plt.subplots(1,2, figsize=(12,6), sharey=True) 
z = sns.violinplot(data=peliculas, x = 'Genero', y='RatingsCriticos',ax = axes[0])
z = sns.boxplot(data=peliculas, x = 'Genero', y='RatingsCriticos',ax = axes[1])
```

```python
#con violinplot vemos el ancho, que indica el numero de ptos que caen en esa área, cosa que en Boxplot no pasa
#muestran lo mismo pero diferente, en boxplot la raya es la media y en violin es el puntico blanco
```

```python
f, axes = plt.subplots(1,2, figsize=(12,6), sharey=True) 

w = sns.boxplot(data=peliculas[peliculas.Genero == 'Drama'], x ='Año', y ='RatingsCriticos', ax=axes[0])
z = sns.violinplot (data=peliculas[peliculas.Genero == 'Drama'], x ='Año', y ='RatingsCriticos', ax=axes[1])
```
