```python
#partimos de la def para gráficos que teniamos de antes, 
def mi_grafico(matriz,lista_jugadores=jugadores):  #añadimos un campo default,(siempre el ultimo tiene que ir)
    col = {"KobeBryant":"black","JoeJohnson":"red","LeBronJames":"green","CarmeloAnthony":"blue","DwightHoward":"red","ChrisBosh":"black","ChrisPaul":"red","KevinDurant":"green","DerrickRose":"blue","DwayneWade":"pink"}
    mar = {"KobeBryant":"s","JoeJohnson":"o","LeBronJames":"d","CarmeloAnthony":"^","DwightHoward":".","ChrisBosh":"<","ChrisPaul":">","KevinDurant":"s","DerrickRose":"o","DwayneWade":"o"}

    for nombre in lista_jugadores:
        plt.plot(matriz[dic_jug[nombre]], c = col[nombre], marker = mar[nombre], ls="-", ms=7, label = nombre)
    plt.xticks(list(range(0,10)),temporadas, rotation ='vertical') #le damos los datos del eje x para que se muestren
    plt.legend(loc='upper left', bbox_to_anchor=(1,1)) #le estamos ubicando la caja de legend fuera del grafico
    plt.title('Estadisticas')
    plt.grid(True)
    plt.show()
```

```python

mi_grafico(puntos,)
mi_grafico(puntos,["KobeBryant"])
```

```python
#podemos representar incluso operaciones entre matrices
mi_grafico(sueldo/juegos)
```

```python
#vemos en el gráfico las anomalias, ya que esas temporadas no se están jugando y salen los datos disparados
#si vemos las métricas dentro del juego
mi_grafico(minutos_jugados)
mi_grafico(puntos)
```

```python
#también se ven datos alterados por temporadas que no se han jugado y se ven afectadas sus estadísticas.
#por ello se normalizan las métricas para evitar anomalias
mi_grafico(tiros_anotados/juegos)
```

```python
# para interpretar las métricas hay que saber 'qué' datos analizamos y qué representan

mi_grafico(minutos_jugados/juegos)
mi_grafico(juegos,["KobeBryant"])
```
