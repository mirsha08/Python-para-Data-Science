```python
# Vamos a representar UsuariosInternet y TasaNatalidad para buscar patrones en los datos
#primero le pasamos los datos (data=datos) y luego le damos las columnas de x e y (da igual el orden para dar los parámetros)

vis3 = sns.lmplot(data=datos, x='UsuariosInternet', y='TasaNatalidad',fit_reg=False )
```

```python
#CAMBIAR EL TAMAÑO DE LOS MARCADORES
```

```python
vis3 = sns.lmplot(data=datos, x='UsuariosInternet', y='TasaNatalidad',fit_reg=False, hue='GrupoIngresos',aspect=2, markers='d')


#en la info vemos el markers para cambiar el tipo de marcador pero nada para el tamaño
#si vamos al final de la descripción vemos: 
###  {scatter,line}_kws : dictionaries
###    Additional keyword arguments to pass to ``plt.scatter`` and
###    ``plt.plot``.    
###      facet_kws : dict
###    Dictionary of keyword arguments for :class:`FacetGrid`.

#viene a decir que tenemos que usar los diccionarios para más parámetros, es un diccionario de matplotlib así que tenemos que ir a la página https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html
#para ver que valores tenemos para el tamaño.. nos dice que es 's'. s-->float or array-like, **shape** (n, ), optional
```

```python
vis3 = sns.lmplot(data=datos, x='UsuariosInternet', y='TasaNatalidad',fit_reg=False, hue='GrupoIngresos',aspect=2, \
                 scatter_kws={'s':100})


#le pasamos el scatter_kws como es un diccionario el par clave:valor, es 's' según la documentación y el valor es 100
```
