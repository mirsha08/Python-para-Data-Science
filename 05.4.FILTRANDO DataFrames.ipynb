```python
#con todo listo vamos:
#el filtrado de aplica a las filas (no a las columnas, como en excel)
```

```python
datos.UsuariosInternet<2
```

```python
#nos devuelve las filas donde la condición se cumple o no se cumple
```

```python
filtro = datos.UsuariosInternet<2
#creamos un nuevo objeto llamado filtro donde metemos nuestra condición.
```

```python
datos[filtro]
#Conceptualmente esto es como R
#nos enseña las filas que CUMPLEN con el filtro, las True
```

```python
datos.TasaNatalidad>40
```

```python
filtro2 = datos.TasaNatalidad>40
datos[filtro2]
```

```python
#también podemos ponerlo todo junto y funciona igual
datos[datos.TasaNatalidad>40]
```

```python
#FILTROS MÁS INTERESANTES
```

```python
#Combinar filtros 
```

```python
#usamos el operador and pero primero tiene que ir una lista o bien usar el operador & para evaluar ambas condiciones.

#usando &
datos[filtro & filtro2]
```

```python
#si lo ponemos directamente sin usar el objeto filtro y filtro2 (se usan () para separar las condiciones y el operador &)

datos[(datos.TasaNatalidad>40) & (datos.UsuariosInternet<2)]
```

# USANDO .at() y .iat()

son funciones especiales para acceder a eltos de un dataframe

La **.at()** es para etiquetas

La **.iat()** es para ubicación con números enteros, iat usa la indexación, es decir cuenta posiciones

```python
#Accesando a elementos individuales
```

```python
#si ponemos
datos[2,2] #da KeyError
```

```python
datos.head()
```

```python
# las funciones .at() y .iat()
```

```python
datos.iat[3,4]  
#cuenta filas y luego columnas, (recordar indice 0) cuenta los números enteros
```

```python
datos.iat[0,0]
```

```python
#.at() solo ve etiquetas, no como el iat que solo cuenta  (el i de indexación)
datos.at[2,'TasaNatalidad']
```

```python
datos[datos.NombrePais == 'Malta']  #en el filtro de antes de Malta vemos el número de fila de sus datos
```

```python
#sabiendo el nº de fila 116 que es la etiqueta de esa fila , en este ejemplo coincide la etiqueta 116 con el nº de fila
datos.at[116,'UsuariosInternet']
```

```python
#¿Por qué necesitamos .at()?
 #creamos un subset de datos
```

```python
sub10 = datos [: : 10]
```

```python
sub10
```

```python
sub10.iat[10,0]
```

```python
sub10.at[10,'NombrePais']
```

```python
# no da el mismo resultado, por que -->
#iat CUENTA  el numero de indexaciones que tiene el subconjunto sub10 
# .at no le interesa la posición, sino que busca las etiquetas, así que busca la etiqueta 10 con nombre pais
```
